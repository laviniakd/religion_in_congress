```python
def get_party(x):
    if x in bio_dict.keys():
        return bio_dict[x]['party']
    else:
        return 'unknown'
    
    
def get_state(x):
    if x in bio_dict.keys():
        if bio_dict[x]['jobPositions']:
            congress_pos = bio_dict[x]['jobPositions'][0]['congressAffiliation']
            if 'represents' in congress_pos.keys():
                return congress_pos['represents']['regionCode']
        return 'unknown'
    else:
        return 'unknown'
    
    
def church_in_bio(x):
    if x in bio_dict.keys():
        return 'church' in bio_dict[x]['profileText']
    else:
        return 'unknown'
    
def get_dw_nominate(x):
    if x in bio_dict.keys():
        return bio_dict[x]['dw_nominate'][0], bio_dict[x]['dw_nominate'][1]

    return None, None
```

```python
import re

def word_tokenizer(text):
    # Find words, contractions, and punctuation
    tokens = re.findall(r"\b\w+(?:'\w+)?\b|[.,!?;]", text)

    # Handle contractions with apostrophes
    processed_tokens = []
    for token in tokens:
        if "'" in token:
            # Split contractions like "how's" into "how" and "'s"
            parts = re.split(r"(')", token)
            processed_tokens.extend([part for part in parts if part])
        else:
            processed_tokens.append(token)

    return processed_tokens

text = "Hello, how's it going? I can't believe it!"
tokens = word_tokenizer(text)
print(tokens)
```

```python
from collections import Counter
import nltk

count_dict = {k: 0 for k in full_keywords}

def get_lexical_overlap(speech):
    speech = word_tokenizer(speech)
    
    if len(speech) > 0:
        count = 0
        for t in full_keywords:
            inst = speech.count(t)
            count += inst
            count_dict[t] += inst

        return count / len(speech) # roughly normalize
    return 0

def get_lexical_overlap_unique(speech):
    speech = word_tokenizer(speech)
    
    if len(speech) > 0:
        count = 0
        for t in full_keywords:
            if t in speech:
                count += 1

        return count / len(speech) # roughly normalize
    return 0

def get_relig_boolean(speech):
    speech = word_tokenizer(speech)
    
    if len(speech) > 0:
        count = 0
        for t in full_keywords:
            if t in speech:
                return True

    return False
```

```python
import pandas as pd
from sklearn import datasets
import statsmodels.api as sm
from sklearn.preprocessing import StandardScaler
from stargazer.stargazer import Stargazer

relevant_columns = ['is_republican', 'reps_south', 'reps_midwest', 'is_in_senate', 'is_abortion', 'is_immigration', 'is_christianity', 'is_science_technology', 'is_health_insurance', 'year']
scaler = StandardScaler()

reg_df = pd.DataFrame()
for r in relevant_columns:
    reg_df[r] = plot_df[r].astype(int)
reg_df['lex_norm'] = scaler.fit_transform(np.asarray(plot_df['lexical']).reshape(-1, 1))
reg_df.dropna(inplace=True)

print(reg_df.dtypes)
print(reg_df.head())

print("scaled")

est = sm.OLS(endog=reg_df['lex_norm'], exog=sm.add_constant(reg_df[relevant_columns])).fit()
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf

# abortion regression
# fit multilevel model
model = smf.mixedlm("lex_norm ~ is_republican + is_republican * is_abortion + is_abortion + year", reg_df, groups=reg_df["reps_south"], re_formula="1 + year")
result = model.fit()

# print summary
print(result.summary())
```
